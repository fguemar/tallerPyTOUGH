```python
#==== PyTOUGH ==== 
#    geometry
from mulgrids import *
#================= 

# CSV writer
import csv
import math

# calls to system
import os

cwd=os.getcwd()
```

```python
geo=mulgrid().from_gmsh(cwd+'/inputs/miMallaGMSH.msh', [50.]*1, convention=0, atmos_type=2, top_elevation=0., chars = ascii_lowercase, spaces=True, block_order=None)
geo
```

```python
# VORONOI este código escribe los vértices de los triángulos --> malla Voronoi
file=open("losNodosDeMiMallaVoronoi.txt","w")
header = ['ID', 'X(m)', 'Y(m)']
# Create a writer object
wrt=csv.writer(file)
# Write the header
wrt.writerow(header)
# Write the data
alpha=0.667794; x0=347198.0; y0=3033788.0

i=1
for node in geo.nodelist[:]:
    xi=x0+(node.pos[0]*math.cos(alpha)-node.pos[1]*math.sin(alpha))
    yi=y0+(node.pos[1]*math.cos(alpha)+node.pos[0]*math.sin(alpha))
    wrt.writerow([i,format(xi, ".2f"),format(yi, ".2f")])
    i=i+1

file.close()
```

```python
#  DELAUNAY  -  Este código escribe los centros de los triángulos
file=open("losNodosDeMiMallaDelaunay.txt","w")
header = ['ID', 'X(m)', 'Y(m)']
# Create a writer object
wrt=csv.writer(file)
# Write the header
wrt.writerow(header)
# Write the data
alpha=0.667794; x0=347198.0; y0=3033788.0
i=1
for cl in geo.columnlist[:]:
    xi=x0+(cl.centre[0]*math.cos(alpha)-cl.centre[1]*math.sin(alpha))
    yi=y0+(cl.centre[1]*math.cos(alpha)+cl.centre[0]*math.sin(alpha))
    wrt.writerow([i,format(xi, ".2f"),format(yi, ".2f")])
    i=i+1
file.close()
```
