```python
#==== PyTOUGH ==== 
#    geometry
from mulgrids import *
#================= 

from numpy import *

# calls to system
import os

cwd=os.getcwd()
```

```python
#carga las elevaciones del Ã¡rea de estudio dada una malla Delaunay de GMSH
#el archivo elevations_Delaunay.txt se construye con la ayuda de QGis para obtener las elevaciones
elevs=np.loadtxt(cwd+'/inputs/elevations_Delaunay.txt')
#elevs[:,2]
```

```python
min(elevs[:,2])
```

```python
#crea un objeto mulgrid a partir de una malla GMSH
geo=mulgrid().from_gmsh('miMallaGMSH.msh', [250.]*14, convention=0, atmos_type=1, top_elevation=min(elevs[:,2]), chars = ascii_lowercase, spaces=True, block_order=None)
geo
```

```python
for cl in geo.columnlist:
    print(cl.surface)
```

```python
geo.fit_surface(elevs, alpha=0.1, beta=0.1, columns=[], min_columns=[], grid_boundary=False, layer_snap=0.0, silent=False)
```

```python
for cl in geo.columnlist:
    print(cl.surface)
```

```python
geo.write_vtk(filename='3geometry.vtu', arrays=None, wells=False, blockmap={}, surface_snap =0.1)
```
